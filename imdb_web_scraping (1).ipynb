```python
#importing the libraries needed
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
```

```python
#creating empty lists to store movie data
rating=[]
year=[]
gross=[]
title=[]
genre =[]
run_time=[]
user_rating=[]

pages = np.arange(1,1000,100)

#for loop to to extract movies details from IMDB
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc"+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    results=soup.find_all("div",class_='lister-item mode-advanced')

    for res in results:
        rating.append(res.find('span',class_='certificate').text.strip())
        year.append(res.find('span',class_='lister-item-year text-muted unbold').text.strip('()'))
        gross.append(res.find('p', class_='sort-num_votes-visible').text.strip().replace('\n',''))
        title.append(res.find("h3",class_="lister-item-header").text.strip().replace('\n',''))
        genre.append(res.find("span",class_='genre').text.strip())
        run_time.append(res.find("span",class_="runtime").text.strip())
        user_rating.append(res.find("div", class_="inline-block ratings-imdb-rating").text.strip())
```

```python
#Creating dataframe
df1 = pd.DataFrame({'rating':rating,'year':year,'gross':gross,'title':title,'run_time':run_time,'genre':genre,'user_rating':user_rating})
```

```python
df1
```

```python
#Separating gross into three columns(movie_gross,number of votes & imdb ranking)
new = df1['gross'].str.split('|', expand= True)
num_votes = new[0].str.split(':', expand= True)
df1["num_votes"]=num_votes[1]
imdb_ranking = new[2].str.split(':', expand= True)
df1["imdb_ranking"]=imdb_ranking[1]
movie_gross = new[1].str.split(':', expand= True)
df1["movie_gross"]=movie_gross[1]
```

```python
#Remove year & number from title column
new1 = df1['title'].str.split('(', expand= True)
movie_title = new1[0].str.split('.', expand= True)
df1['movie_title']=movie_title[1]
```

```python
#Removing title & gross column
df1.drop(columns=['title', 'gross'],axis=1, inplace = True)
```
